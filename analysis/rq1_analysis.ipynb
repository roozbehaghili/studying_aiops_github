# Statistical Tests

```python
# Normality test: Shapiro-Wilk Test
from numpy.random import seed
from numpy.random import randn
from scipy.stats import shapiro
def shapiro_wilk(d):
    data = d
    # seed the random number generator
    seed(1)
    # normality test
    stat, p = shapiro(data)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print('Sample looks Gaussian (fail to reject H0)')
    else:
        print('Sample does not look Gaussian (reject H0)')

shapiro_wilk(aiops['stars'])
shapiro_wilk(aiops['forks'])
shapiro_wilk(aiops['commits'])
shapiro_wilk(aiops['contributors'])
shapiro_wilk(aiops['releases'])
shapiro_wilk(aiops['pulls'])
shapiro_wilk(aiops['issues'])
shapiro_wilk(aiops['size'])
shapiro_wilk(ml_final['stars'])
shapiro_wilk(ml_final['forks'])
shapiro_wilk(ml_final['commits'])
shapiro_wilk(ml_final['contributors'])
shapiro_wilk(ml_final['releases'])
shapiro_wilk(ml_final['pulls'])
shapiro_wilk(ml_final['issues'])
shapiro_wilk(ml_final['size'])
shapiro_wilk(general_final['stars'])
shapiro_wilk(general_final['forks'])
shapiro_wilk(general_final['commits'])
shapiro_wilk(general_final['contributors'])
shapiro_wilk(general_final['releases'])
shapiro_wilk(general_final['pulls'])
shapiro_wilk(general_final['issues'])
shapiro_wilk(general_final['size'])
```

The distributions are NOT normall. Hence, we use non-parametric tests.

```python
# Statistical test: Mann-Whitney U Test (Wilcoxon rank-sum test)
from scipy.stats import mannwhitneyu
def whitney(d1, d2, par):
    stat, p = mannwhitneyu(d1, d2, alternative='two-sided')
    print('stat=%.2f, p=%.2f' % (stat, p))
    if p > 0.05:
        print(par, ': Probably same distribution')
    else:
        print(par, ': Probably DIFFERENT distributions')
whitney(aiops['stars'], ml_final['stars'], 'stars')
whitney(aiops['stars'], general_final['stars'], 'stars')
whitney(aiops['forks'], ml_final['forks'], 'forks')
whitney(aiops['forks'], general_final['forks'], 'forks')
whitney(aiops['commits'], ml_final['commits'], 'commits')
whitney(aiops['commits'], general_final['commits'], 'commits')
whitney(aiops['contributors'], ml_final['contributors'], 'contributors')
whitney(aiops['contributors'], general_final['contributors'], 'contributors')
whitney(aiops['releases'], ml_final['releases'], 'releases')
whitney(aiops['releases'], general_final['releases'], 'releases')
whitney(aiops['pulls'], ml_final['pulls'], 'pulls')
whitney(aiops['pulls'], general_final['pulls'], 'pulls')
whitney(aiops['issues'], ml_final['issues'], 'issues')
whitney(aiops['issues'], general_final['issues'], 'issues')
whitney(aiops['size'], ml_final['size'], 'size')
whitney(aiops['size'], general_final['size'], 'size')
```
